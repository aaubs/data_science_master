## Introduction
    Gradient Descent is a fundamental optimization algorithm used in machine learning to minimize a function. It's
    particularly useful in training machine learning models. In this tutorial, we will explore Gradient Descent,
    focusing on how to minimize the Mean Squared Error (MSE) in a linear regression problem.
    


## 1. What is Gradient Descent?
    Gradient Descent is an optimization algorithm that aims to find the minimum of a function. It iteratively moves
    in the direction of the steepest decrease in the function, which is the opposite direction of the gradient of the
    function.
    

The derivative and the gradient are both fundamental concepts in calculus, each describing how functions change, but they are applied in distinct contexts:

**Derivative of a Function**: The derivative is relevant to single-variable functions. Consider a function `f(x)`; its derivative, denoted as `f'(x)` or `df/dx`, signifies the rate of change of the function's value in relation to variations in `x`. Geometrically, for a curve represented by `y = f(x)`, the derivative at a particular point determines the slope of the tangent line at that point on the curve. Essentially, the derivative indicates how fast or slow the function value is rising or falling at each point.

**Gradient of a Function**: The gradient concept is an extension of the derivative for functions with multiple variables. For a function `f(x, y, z, ...)`, the gradient, represented as `∇f` or `grad f`, is a vector that points in the direction where the function increases most steeply. In a three-dimensional space, it is expressed as `∇f = ( ∂f/∂x, ∂f/∂y, ∂f/∂z )`. Each component of this gradient vector is the partial derivative of the function with respect to one of its variables, showing how the function changes as that particular variable changes, while other variables are held constant.


# Gradient and Gradient Descent

## Gradient

In mathematics and optimization, a **gradient** of a function is a vector consisting of the partial derivatives of that function with respect to each variable. It represents:

- **For Single Variable**: The slope of the function at a specific point.
- **For Multiple Variables**: The slope in each dimension.

## Gradient Descent

**Gradient Descent** is an optimization algorithm used to find the minimum value of a function. It involves:

1. **Starting Point**
   - Begin at a random point on the function.

2. **Calculating the Gradient**
   - Determine the gradient at this point, showing the direction of steepest ascent.

3. **Moving in Opposite Direction**
   - To find the minimum, move in the opposite direction of the gradient.

4. **Step Size (Learning Rate)**
   - The size of the step, determined by the learning rate, is critical. A high learning rate might overshoot the minimum, while a low rate may cause slow convergence.

5. **Iterative Process**
   - Repeat the process iteratively, moving closer to the minimum in each step.

6. **Convergence**
   - The algorithm stops when it converges, indicating the minimum is reached (or nearly reached).

## Application in Machine Learning

Gradient Descent is extensively used in machine learning for training models by minimizing a loss function, which measures the difference between the model's predictions and the actual data.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function and its derivative as Python functions
def f(x):
    return x**2 + 3*x + 2

def f_prime(x):
    return 2*x + 3

# Generate x values
x_values = np.linspace(-10, 10, 400)

# Generate y values for the function and its derivative
y_values = f(x_values)
y_prime_values = f_prime(x_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = $x^2 + 3x + 2$')
plt.plot(x_values, y_prime_values, label="f'(x) = $2x + 3$", linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and its Derivative')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```


## 2. What is Mean Squared Error (MSE)?
    Mean Squared Error is a common loss function used in regression problems. It calculates the average of the squares
    of the errors between the predicted values and the actual values.
$MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2$

    where:
- \( n \) is the number of data points or samples.
- \( $y_i$ \) is the actual value of the \( $i^{th}$ \) data point.
- \( $\hat{y}_i$) is the predicted value of the \( $i^{th}$ ) data point.
    

## 3. The Gradient Descent Algorithm
    The Gradient Descent algorithm involves several steps:
    
### 3.1 Initialization
    Choose a starting point for the parameter(s) you are trying to optimize.
    
### 3.2 Compute the Gradient
    Calculate the gradient of the loss function with respect to the parameter(s).
    
### 3.3 Update the Parameter(s)
    Update the parameter(s) in the direction opposite to the gradient.
    
### 3.4 Repeat
    Repeat steps 2 and 3 until convergence.
    


## 4. Learning Rate
    The learning rate $(alpha)$ is a crucial hyperparameter in Gradient Descent. It determines the size of the steps
    we take towards the minimum. If it's too high, the algorithm might overshoot the minimum. If it's too low, the
    algorithm will take a very long time to converge.
    


## 5. Types of Gradient Descent
    There are three main types of Gradient Descent:
    
### 5.1 Batch Gradient Descent
    Uses the entire dataset to compute the gradient of the cost function.
    
### 5.2 Stochastic Gradient Descent
    Uses only one data point at a time to compute the gradient of the cost function.
    
### 5.3 Mini-batch Gradient Descent
    Uses a subset of the data to compute the gradient of the cost function.
    



## 6. Practical Example with Python
    Let's go through a simple example to demonstrate how gradient descent works, particularly for minimizing the
    Mean Squared Error (MSE) in a linear regression problem.
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Example Data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])

# Initialize parameters
m = 0
b = 0

# Learning rate
alpha = 0.01

# Number of iterations
iterations = 5

# Number of data points
n = len(X)

# For storing values for visualization
m_values = []
b_values = []
loss_values = []

# Gradient Descent Loop
for _ in range(iterations):
    # Compute the Loss (Mean Squared Error)
    Y_pred = m * X + b
    loss = (1/n) * sum((Y - Y_pred)**2)

    # Compute the Gradients
    dm = -(2/n) * sum(X * (Y - Y_pred))
    db = -(2/n) * sum(Y - Y_pred)

    # Update Parameters
    m = m - alpha * dm
    b = b - alpha * db

    # Store values for visualization
    m_values.append(m)
    b_values.append(b)
    loss_values.append(loss)

    # Visualization of the results
    plt.scatter(X, Y, color='blue')
    plt.plot(X, Y_pred, color='red')
    plt.title('Final Fit')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

Here's a list of some important derivative formulas for various functions:

## Power Rule

Power Rule for Derivatives:

$d/dx(x^n) = nx^{n-1}$

This applies to any real number n, where x^n is a power of x.

## Constant Rule

Constant Rule for Derivatives:

$d/dx(c) = 0$

This states that the derivative of a constant c is zero.

## Constant Multiple Rule

Constant Multiple Rule for Derivatives:

$d/dx(cf(x)) = c \cdot d/dx f(x)$

Here, c is a constant and f(x) is a function.

## Sum Rule

Sum Rule for Derivatives:

$d/dx(f(x) + g(x)) = d/dx f(x) + d/dx g(x)$

The derivative of the sum of two functions is the sum of their derivatives.

## Difference Rule

Difference Rule for Derivatives:

$d/dx(f(x) - g(x)) = d/dx f(x) - d/dx g(x)$

Similar to the sum rule, but for the difference of two functions.

## Product Rule

Product Rule for Derivatives:

$d/dx(f(x)g(x)) = f'(x)g(x) + f(x)g'(x)$

The derivative of a product of two functions.

## Quotient Rule

Quotient Rule for Derivatives:

$d/dx(f(x)/g(x)) = (f'(x)g(x) - f(x)g'(x)) / [g(x)]^2$

This is used for the division of two functions.

## Chain Rule

Chain Rule for Derivatives:

$d/dx f(g(x)) = f'(g(x))g'(x)$

Used when a function is composed of another function.

## Derivative of Exponential Functions

Exponential Function Derivatives:

$d/dx e^x = e^x$

$d/dx a^x = a^x \ln(a)$ (where a > 0 and a ≠ 1)

## Derivative of Logarithmic Functions

Logarithmic Function Derivatives:

$d/dx \ln(x) = 1/x$

$d/dx \log_a(x) = 1 / (x \ln(a))$ (where a > 0 and a ≠ 1)

## Derivatives of Trigonometric Functions

Trigonometric Function Derivatives:

$d/dx \sin(x) = \cos(x)$

$d/dx \cos(x) = -\sin(x)$

$d/dx \tan(x) = \sec^2(x)$

## Derivatives of Inverse Trigonometric Functions

Inverse Trigonometric Function Derivatives:

$d/dx \arcsin(x) = 1 / \sqrt{1-x^2}$

$d/dx \arccos(x) = -1 / \sqrt{1-x^2}$

$d/dx \arctan(x) = 1 / (1+x^2)$
